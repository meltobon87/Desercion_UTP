***Pycaret***

```python
from pycaret import classification
s = classification.setup(datos, target = "permanencia", session_id = 1234,
                         normalize = True, fix_imbalance=True)
```

```python
classification.get_config('X_transformed')
```

```python
mejor_modelo = classification.compare_models(sort='AUC')
```

Reg. Log

```python
modelo_RL= classification.create_model('lr')
modelo_RL
```

## Análisis modelo_RL

```python
classification.evaluate_model(modelo_RL)
```

```python
X_test = classification.get_config('X_test')
y_test = classification.get_config('y_test')

X_train = classification.get_config('X_train')
y_train = classification.get_config('y_train')

Xy_test = pd.concat([X_test, y_test], axis=1)
Xy_train = pd.concat([X_train, y_train], axis=1)
```

## Predicciones

```python
data_test_pred = classification.predict_model(modelo_RL,raw_score=True)
data_test_pred
```

```python
data_pred = classification.predict_model(modelo_RL, data=datos, raw_score=True)
data_pred
```

```python
import sklearn.metrics
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

cm = confusion_matrix(data_pred['permanencia'], data_pred['prediction_label'])
ConfusionMatrixDisplay(cm, display_labels=['Permanece', 'Deserta']).plot()
```

```python
import bctools as bc

threshold_step = 0.05
curve_type = 'kde' #'kde' is the default value, can also be set to 'normal'
bc.predicted_proba_density_curve_plot(true_y = data_test_pred['permanencia'],
                                      predicted_proba = data_test_pred['prediction_score_1'],
                                      threshold_step = threshold_step,
                                      curve_type = curve_type)
```

```python
# set params for the train dataset
threshold_step = 0.025
optimize_threshold = None
currency = '$'
# It takes as input, for each class, a float or a list of floats.
# Lists must have coherent lenghts
# Ejemplo: FP: cuesta $10 equivocarnos en pacientes que realmente no estan enfermos,
# FN: Cuesta $5 equivocarnos en pacientes que estan realmente enfermos (nosotros decimos que estan sanos)
# TP: Pacientes que están realmente enfermos y en el modelo decimos que están enfermos, acertar tiene un beneficio, por lo tanto se le da un valor negativo, para este caso $-15
train_cost_dict = bc.get_cost_dict(TN = 0, FP = 10, FN = 5, TP = -15)

results = bc.confusion_matrix_plot(
    true_y = data_test_pred['permanencia'],
    predicted_proba = data_test_pred['prediction_score_1'],
    threshold_step = threshold_step,
    cost_dict = train_cost_dict,
    # optimize_threshold = optimize_threshold,
    currency = currency,
    # random_state = 123,
    title = 'Interactive Confusion Matrix for the Training Set');
```
